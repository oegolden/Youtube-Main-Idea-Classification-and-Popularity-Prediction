Classify into Not Popular, Semi-Popular, Popular, Viral, and Mega-Viral

```python
import math
def viewClassifier(num):
    if num < math.pow(10,5) -  1:
        return("Not Popular")
    elif num < math.pow(10,6) - 1:
        return("Semi-Popular")
    elif num < math.pow(10,6) * 5:
        return("Popular")
    elif num < math.pow(10,7) * 5:
        return("Viral")
    elif num < math.pow(10,8) - 1:
        return("Mega-Viral")
    elif num < math.pow(10,9) * 5:
        return("Historically-Viral")


df['popularity'] = df.apply(lambda x: viewClassifier(x['views']), axis=1)
df
```

Use Spacy word similarity on all the trags, to find the tag with the highest similarity to the other tags, and then use that tag as the label, and the title as the example.


```python
# First I need to be able to break up the tags into a list of the seperate tags
# And that list needs to be broken into seperate tokens
def token_creator(row):
    tags = df.iloc[row]["tags"].split("|")
    tags = map(lambda x: x.replace('"',''), tags)
    tag_tokens = []
    for x in tags:
        tag_tokens.append(nlp(x))
    indiv_tokens = []
    for x in tag_tokens:
        for token in x:
            indiv_tokens.append(token)
    for token in indiv_tokens:
        if (not (token.pos_ == "NOUN")) and (not (token.pos_ == "VERB")) and (not (token.pos_ == "PROPN")) and (not (token.pos_ == "PRON")):
            indiv_tokens.remove(token)
    for token in indiv_tokens:
        if (not token.has_vector):
            indiv_tokens.remove(token)
    indiv_tokens = map(lambda x: nlp(x.lemma_), indiv_tokens)
    indiv_tokens = list(indiv_tokens)
    return(indiv_tokens)
indiv_tokens = token_creator(24511)
```

Now that I have the individual tokens, I can run a similarity check on them and divide the token's into the most appearing ones, and similar ones i.e. animal would go into puppies, because puppies appears more than animals.



Another strategy to decide the tag that best describes the video, could be to pick the tags with the highest average similarity to the other tags.

```python
def average_similarity_of_list(tokens):
    if len(tokens) == 1:
        return(dict({tokens[0]:1.0}))
    token_averages = {}
    for x in tokens:
        if (x.text not in list(map(lambda y: y.text, token_averages))):
            sum = 0
            for token in tokens:
                if (x.text != token.text):
                    sum += x.similarity(token)
            token_averages[x] = sum
    token_averages = {k: v/len(token_averages) for k, v in token_averages.items()}
    return(dict(sorted(token_averages.items(), key=lambda item: item[1])))
```

```python
averages = average_similarity_of_list(indiv_tokens)
averages
```

Now that I have a function for what tags are most indicative to what all the tags mean (similarity function isn't exactly that, but its the best I can do for word similarity/definition similarity) I can create a dictionary of the top 5 tags(if there are 5 tags) and the title as examples for those.

```python
def classification(index):
    print(str(index) + " out of " + str(len(df)))
    indiv_tokens = token_creator(index)
    averages = average_similarity_of_list(indiv_tokens)
    i = 0
    main_ideas = []
    length = len(list(averages.keys())) - 1
    while i < 5 and i < length:
        main_ideas.append(list(averages.keys())[length - i])
        i+= 1
    return(list(main_ideas))

df['classes'] = df.apply(lambda x: classification(x["index"]), axis=1)
df
```

We can see on average each key has ~5 titles, meaning that on average any title shares a main idea with 4 other titles, which does not seem like nearly enough examples for a classifier to be extremely effective; nevertheless we should prove this hypothesis.

```python
classifier_df = pd.DataFrame()

for key, value in data.items():
    for x in value:
       tempdf = pd.DataFrame({'title': [x],
        'tag': [key]})
       classifier_df = classifier_df.append(tempdf, ignore_index=True)
classifier_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
X = classifier_df.title
y = classifier_df.tag
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python

from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer

# We First Vectorize both the tags and the titles with CountVectorizer
# We then use the TFidfTransformer to reduce the weight of tokens that appear more frequently
# And then calculate the probability of the tag belonging to that system with the MultionamialNB classifier
nb = Pipeline([('vect', CountVectorizer()),
               ('tfidf', TfidfTransformer()),
               ('clf', MultinomialNB()),
              ])
nb.fit(X_train, y_train)

from sklearn.metrics import classification_report
y_pred = nb.predict(X_test)

print('accuracy %s' % accuracy_score(y_pred, y_test))
```
